```python
metadata = pd.read_csv("/content/drive/Shareddrives/MLCB_project_dataset/schic_snm3cseq/liu_ecker_2023/CEMBA.m3C.Metadata.csv")
full_directory = pd.read_csv("/content/drive/Shareddrives/MLCB_project_dataset/schic_snm3cseq/data/filelist.txt", sep = "\t", header= None).iloc[:,0].to_list()

pattern = r"\/([A-Za-z0-9_]+-\d+-[A-Z]+\d+-[A-Z]+\d+)\.3C\.contact"

# Extract cell IDs
cleaned_cell_id = []
for path in full_directory:
    match = re.search(pattern, path)
    if match:
        cleaned_cell_id.append(match.group(1))

training_data_meta = pd.DataFrame(cleaned_cell_id, columns = ["cell"]).merge(metadata, on = "cell", how = "left")
training_data_meta
```

```python
# Load metadata from mC
import re
pattern = r"^[^_]+"
#pattern = r'^([^_]+_[^_]+)'
mc_cell_group = mc_meta.CellGroup.to_list()
mc_meta_cellgroup = [re.match(pattern, cell_group).group() for cell_group in mc_cell_group]
mc_meta["cellgrp"] = mc_meta_cellgroup
merged_cellgroup = merged.CellGroup.to_list()
merged_cellgroup = [re.match(pattern, cell_group).group() for cell_group in merged_cellgroup]
merged["cellgrp"] = merged_cellgroup
merged
```

```python
cellgrp_to_class = mc_meta.groupby('cellgrp')['Class'].apply(lambda x: list(set(x))).to_dict()
cellgrp_to_class
```

```python
for idx, (index, row) in enumerate(merged.iterrows()):
  annot = row['Cell Subclasses Label \n(From AIBS scRNA-seq (10X) Annotation)']
  if pd.isnull(annot):
    merged.loc[index, 'Cell Subclasses Label \n(From AIBS scRNA-seq (10X) Annotation)'] = cellgrp_to_class[row["cellgrp"]][0]

merged
```

```python
def classify_cell_group(group):
    if ("Gaba" in group) or ("GABA" in group):
        return "Inhibitory Neuron"
    elif ("Glut" in group) or ("GLUT" in group):
        return "Excitatory Neuron"
    elif "NN" in group or "Vascular" in group or "OB-STR-CTX Inh IMN" in group:
        return "Non-Neuronal"


cell_group_mapping = {group: classify_cell_group(group) for group in merged['Cell Subclasses Label \n(From AIBS scRNA-seq (10X) Annotation)'].unique()}


plotting_grps = []
for idx, (index, row) in enumerate(merged.iterrows()):
  plotting_grps.append(classify_cell_group(row['Cell Subclasses Label \n(From AIBS scRNA-seq (10X) Annotation)']))
plotting_grps
```
